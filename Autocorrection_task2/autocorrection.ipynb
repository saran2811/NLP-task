```python
from textblob import TextBlob
import pandas as pd
```

```python
df=pd.read_excel("/content/mispelled words.xlsx")
df
```

```python
def auto_correction(Word,i):
  word=TextBlob(Word)
  cor_word=word.correct()
  df['Corrected_words'][i]=cor_word
```

```python
df['Corrected_words']=0*len(df)
for i in range(len(df)):
  auto_correction(df['Misspelled Words'][i],i)
```
